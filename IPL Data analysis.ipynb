Total number of match played during each year.

```python
match_per_year=match_data.groupby('year')['id'].count().reset_index().rename(columns={'id':'matches'})

match_per_year
```

```python
sns.countplot(match_data['year'])
plt.xticks(rotation=45,fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel('Year',fontsize=10)
plt.ylabel('Count',fontsize=10)
plt.title('Toatal no of matches played each year',fontsize=10,fontweight='bold')
```

```python
season_data=match_data[['id','year']].merge(ball_data,on=['id','id'],how='left').drop('id',axis=1)
season_data
```

Total runs scored during each season.

```python
season=season_data.groupby('year')['total_runs'].sum().reset_index()
```

```python
p=season.set_index(season['year'])
```

```python
sns.lineplot(data=p)
plt.show()
```

```python
df_player = ball_data[ball_data['batsman']=='SK Raina']
df_player
```

```python
df_player['dismissal_kind'].value_counts().plot.pie(autopct='%1.1f%%')
```

```python
def count(df_player,runs):
    return len(df_player[df_player['batsman_runs']==runs])*runs
```

```python
print('Runs scored by suresh raina from 1s', count(df_player,1))
print('Runs scored by suresh raina  from 2s', count(df_player,2))
print('Runs scored by suresh raina  from 3s', count(df_player,3))
print('Runs scored by suresh raina  from 4s', count(df_player,4))
print('Runs scored by suresh raina from 6s', count(df_player,6))
```
